```python
import pandas as pd
import nltk
from gensim.models import word2vec
```

```python
corpus=[
    "I like eating apple",
    "An apple a day keeps the doctor away",
    "Never compare an apple to an orange",
    "I prefer scikit-learn to orange"
]
```

```python
tokenize_corpus=[sentence.lower().split() for sentence in corpus]
tokenize_corpus
```

```python
model=word2vec.Word2Vec(tokenize_corpus,vector_size=100,window=3,min_count=1,sg=0)
model.save("word2vec.model")
vector = model.wv["apple"]
print(vector)
```

```python
similar_words=model.wv.most_similar("apple")
for word,similar in similar_words:
    print(f"{word} : {similar}")
```
